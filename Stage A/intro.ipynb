The Food Balance Sheets (FBS) dataset provides essential information on a country’s food system.

```python
# package(s) needed
import pandas as pd
```

### **QUICK LOOK AT DATA**

```python
df = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin-1')
df.head()
```

### **WHAT IS THE TOTAL NUMBER OF UNIQUE COUNTRIES IN THE DATASET?**

```python
df['Area'].nunique()
```

### **PERFORM A GROUPBY OPERATION ON ‘ELEMENT’. WHAT YEAR HAS THE HIGHEST SUM OF STOCK VARIATION?**

```python
df.groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum()
```

```python
df.groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum()\
        .reset_index().query('Element == "Stock Variation"')
```

> The result above shows that the year 2014 has the highest sum of Stock Variation.

### **PERFORM A GROUPBY OPERATION ON ‘ELEMENT’. WHAT IS THE TOTAL NUMBER OF THE SUM OF PROCESSING IN 2017?**

```python
df.groupby('Element')[['Y2017']].sum()
```

```python
df.groupby('Element')[['Y2017']].sum().reset_index().query('Element == "Processing"')
```

> The total number of the sum of processing in 2017 is 292836.0.

### **SELECT COLUMNS ‘Y2017’ AND ‘AREA’, PERFORM A GROUPBY OPERATION ON ‘AREA’.  WHICH OF THESE AREAS HAD THE 7TH LOWEST SUM IN 2017?**

```python
df.groupby('Area')[['Y2017', 'Area']].sum().sort_values(by='Y2017')[:10]
```

> Using the code above was concatenating the Area, so I used the code below to get a better format for the output.

```python
df.groupby('Area')[['Y2017']].sum().sort_values(by='Y2017')[:10]
```

> The Area with the 7th lowest sum in year 2017 is Guinea-Bissau.

### **WHAT IS THE TOTAL SUM OF WINE PRODUCED IN 2015 AND 2018 RESPECTIVELY?**

```python
df.groupby('Item')[['Y2015', 'Y2018']].sum()
```

```python
df.groupby('Item')[['Y2015', 'Y2018']].sum()\
    .reset_index().query('Item == "Wine"')
```

> The total sum of Wine produced in 2015 and 2018 respectively are 4251.81 and 4039.32.

### **SELECT COLUMNS ‘Y2017’ AND ‘AREA’, PERFORM A GROUPBY OPERATION ON ‘AREA’.  WHICH OF THESE AREAS HAD THE HIGHEST SUM IN 2017?**

```python
df.groupby('Area')[['Y2017', 'Area']].sum().sort_values(by='Y2017', ascending=False)[:10]
```

> Using the code above was concatenating the Area, so I used the code below to get a better format for the output and I reduced the output to 10 rows.

```python
df.groupby('Area')[['Y2017']].sum().sort_values(by='Y2017', ascending=False)[:10]
```

> The Area with the highest sum in 2017 is Nigeria.

### **WHAT IS THE MEAN AND STANDARD DEVIATION ACROSS THE WHOLE DATASET FOR THE YEAR 2017 TO 2 DECIMAL PLACES?**

```python
df['Y2017'].describe()
```

> The mean and the standard deviation across the whole dataset for the year 2017 are 140.92 and 1671.86 respectively.

### **WHICH YEAR HAD THE LEAST CORRELATION WITH ‘ELEMENT CODE’?**

```python
df.select_dtypes(include='number').corr()
```

> The year with the least correlation with Element Code is 2016.

### **WHAT IS THE TOTAL NUMBER AND PERCENTAGE OF MISSING DATA IN 2014 TO 3 DECIMAL PLACES?**

```python
df['Y2014'].isnull().sum()
```

```python
round((df['Y2014'].isnull().sum() / df.shape[0]) * 100, 3)
```

> The total number and percentage of missing data in 2014 are 1589 and 2.607 3dp respectively.

### **WHAT IS THE TOTAL PROTEIN SUPPLY QUANTITY IN MADAGASCAR IN 2015?**

```python
df.query('Element == "Protein supply quantity (g/capita/day)" and Area == "Madagascar"')\
    ['Y2015'].sum()
```
